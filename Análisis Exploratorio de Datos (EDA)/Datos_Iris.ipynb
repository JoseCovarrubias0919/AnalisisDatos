Se importan los recursos requerido para el algoritmo

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Mandamos llamar al archivo csv y vemos su contenido

```python
iris = pd.read_csv("CSV/Iris.csv")
#Se manda llamar las cinco primeras filas del csv
print(iris.head(5))
#Se despliega la informacion de los tipos de dato de cada columna
print(iris.info())
```

```python
# Visualizacin de los datos
sns.pairplot(iris, hue='Species', markers=["o", "s", "D"])
plt.show()
```

Analisis de los datos

```python
# Distribucin de la longitud del spalo
sns.histplot(iris['SepalLengthCm'], kde=True)
plt.xlabel('Sepal Length (cm)')
plt.title('Distribution of Sepal Length')
plt.show()
```

```python
# Boxplot de la longitud del spalo por especie
sns.boxplot(x='Species', y='SepalLengthCm', data=iris)
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.title('Sepal Length by Species')
plt.show()
```

```python
# Scatter plot de Sepal Length vs Sepal Width
sns.scatterplot(data=iris, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
# Scatter plot de Petal Length vs Petal Width
sns.scatterplot(data=iris, x='PetalLengthCm', y='PetalWidthCm', hue='Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs Petal Width')
plt.show()
```
